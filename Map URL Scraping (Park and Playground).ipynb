```python
import requests

def get_parks_and_playgrounds(city):
    url = "https://overpass-api.de/api/interpreter"
    query = f"""
    [out:json];
    area["name"="{city}"]->.searchArea;
    (
      node["leisure"="park"](area.searchArea);
      node["leisure"="playground"](area.searchArea);
      way["leisure"="park"](area.searchArea);
      way["leisure"="playground"](area.searchArea);
    );
    out center;
    """
    
    response = requests.get(url, params={"data": query})
    if response.status_code == 200:
        data = response.json()
        parks_and_playgrounds = []
        for element in data["elements"]:
            name = element.get("tags", {}).get("name", "Unnamed")
            lat = element.get("lat") or element.get("center", {}).get("lat")
            lon = element.get("lon") or element.get("center", {}).get("lon")
            parks_and_playgrounds.append({"name": name, "lat": lat, "lon": lon})
        return parks_and_playgrounds
    else:
        print(f"Error: {response.status_code}")
        return None

# Example Usage
city = "Ahmedabad"
results = get_parks_and_playgrounds(city)
if results:
    print(f"Parks and Playgrounds in {city}:")
    for idx, place in enumerate(results, 1):
        print(f"{idx}. {place['name']} (Lat: {place['lat']}, Lon: {place['lon']})")
```
